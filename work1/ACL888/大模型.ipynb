机器学习分为3种方式：有监督学习，自监督学习，强化学习。其本质是通过不断调整参数权重使模型拟合程度不断提升，损失函数达到最小。

基本术语：
1.数据：训练检验所用的样本集，一个事物的特征集合；
2，模型：显示映射关系的函数（我们要做的就是不断改进模型使其产生的结果更接近真实值）；
3.标签：类似于答案，是数据对于的真实结果；
4.特征：数据的属性，是模型输入的变量；
5，环境：数据存在于环境中，通过模型来改变环境是我们人类的目标。


大模型的基本训练流程：
1，找到合适的数据；
2，订好合适的模型结构如线性结构，softmax结构，循环神经网络，transformer结构（现在大模型采用）等；
3，清洗数据，去除无关信息，如序列号等，避免数据污染，导致模型训练偏差；
4，将数据随机打乱，划分训练集，验证集，测试集；
5，选择合适的损失函数，如交叉熵损失函数（softmax），均方误差损失函数等(线性回归）；
6，展开多维数据，进行矩阵运算，提升计算效率；
7，设定超参数，如学习率，batch_size，epoch等；
8，设定优化器，如梯度下降，Adam等；
9，设置其他内容，如可视化；
10，开始训练，并根据训练集第一次结果对超参数进行调整（努力防止过拟合）（期间采用反向传播算法）；
11，关闭反向传播，关闭优化器，使用测试集进行最终测试，评估模型效果。（不要让模型提前看到答案！！！）
12，分析泛化性，拟合性，评估模型。

1.有监督学习：人类给予数据，标签，特征以此来训练模型，相当于给模型答案让其找规律；
2，自监督学习：人类给予数据让模型自己来总结特征及标签以此来训练自己，相当于考试完后家长或老师扔给你题目让你自己找错因，答案并分析（感觉比较贴切）；
3，强化学习：前二者都只是观察，分析环境，输入数据也是死板的，而强化学习与他们区别在于其会根据以往分析得出的进行输出，并观察反思总结自己输出对外界环境所带来的影响，并将其作为输入再一次训练自身，不断反复，以此来提高自身性能，
其从观察者变为了参与者。
