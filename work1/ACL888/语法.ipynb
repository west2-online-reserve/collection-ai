函数
1.Lambda匿名函数
 一次性的简单函数多用于复杂计算，高阶函数的参数或返回值
 不能包含多语句，只可用单行表达式

2.Decorator.装饰器
 2.1 一种高阶函数输入一个函数其会在不改变函数名称，函数签名等情况下对其进行修改增强
 2.2 内部定义的新函数可以接受原函数所有参数可以添加任意代码
 2.3 用@放在被装饰函数的上方。

面向对象
1.类
 1.1 用class定义
 1.2 用__init__初始化（self，各种属性
 1.3 方法是指类中运用属性所编的函数（很想）
 1.4 子类可用super（）来继承父类属性及方法
 1.5 封装，可以对部分方法属性封装来控制访问级别


2.Magic Method 魔法方法
 类中被定义好的有着特殊功能的方法，特征是__x___
 如__init__(初始化),__str__(输出)，__add__(加入)

3.OOP面向对象编程思想
  1.封装：把数据 & 行为放在同一个类里，隐藏内部实现细节
  2.继承：子类继承父类的属性和方法，实现代码复用
  3.多态：不同类对象对同一个信息响应，提升灵活性
  4.单一职责：每一类只负责一个抽象概念
  5.组合：一个类的对象作为另一个类对象成员，灵活与
  于继承形成整体-部分关系
  6.抽象：具体事物关注共同特征选择忽略不同细节，将万物抽象为类。

文本处理
1.re正则表达式
 1.1用于匹配，查找，编辑符合特定要求的文本字符串
 1.2 .匹配任意字符，*匹配前一个字符0次或多次，+匹配前一个字符1次或多次，?匹配前一个字符0次或1次，同时也可切换贪婪模式和懒惰模式
 1.3 ^匹配字符串开头，$匹配字符串结尾，\b匹配单词边界
 1.4 []定义字符类，|表示或，()表示分组
 1.5 贪婪模式与懒惰模式
  贪婪模式：尽可能多地匹配字符
  懒惰模式：尽可能少地匹配字符


代码格式
1.列表推导式
 1.1 用于生成列表的简洁语法
 1.2 语法：[expression for item in iterable if condition]
 1.3 可包含多个for和if子句实现复杂逻辑
 1.4 提高代码可读性和执行效率
 1.5但是过度复杂会降低代码可读性


2.Type Hint（类型注释）（非强制）
 2.1 用于指示变量，函数参数和返回值的预期数据类型
 2.2 对于参数和返回值使用冒号和箭头进行注释
 2.3 对于变量使用冒号进行注释
 2.4 提高代码可读性和维护性，在大型项目卓有帮助
