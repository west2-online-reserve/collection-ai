1. 自监督学习

+ 自监督学习是AI学习的第一步。在此过程中，模型会阅读大量的公开文本资料，并从中学习语言的基本规律。
+ 在此阶段，AI并不是由人类一步步教导，而是自己预测在文字接龙时下一个字可能是什么，从而逐渐掌握语言的结构、语义和逻辑关系。即“自我阅读”

2. 监督式学习

+ 在拥有语言基础之后，AI会接受人类提供的高质量问答示例进行训练，即学习一个合适的回答方式。
+ 通过这一过程，可以让模型更加准确地理解用户的意图并组织更加自然的语言，同时避免明显的错误。

3. 人类反馈强化学习

+ 这一过程主要是通过人类的反馈对模型进行优化。
+ 人类可以在此阶段对AI的多个回答进行评估并将其再次喂给AI，AI通过强化学习算法不断调整，使得自己的答案更倾向于人类偏好的内容。
