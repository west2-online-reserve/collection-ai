# 1.关于列表(list)：

+ 一般使用"[ ]"表示列表。列表可用于储存一组有特定排列方式的数据，可以将任何东西加入列表，且列表中的元素之间除了顺序以外没有任何关系。列表可以被遍历。

```python
bicycles = ['trek','cannondale','redline','specialized']
print(bicycles)
```

+ 访问列表中的元素时，可以使用列表名加索引的形式，列表的第一个元素的索引是0。列表的索引也可以反着看，但是最后一个元素的索引是-1。

```python
print(bicycles[0])
```

+ 在列表中修改元素时，可以直接使用列表名加索引，并令其为一个新的值。

```python
bicycles = ['trek','cannondale','redline','specialized']
bicycles[0] = 'ducati'
print(bicycles)
```

+ 列表的用法

1. 首字母大写：

```python
bicycles = ['trek','cannondale','redline','specialized']
message = f'My first bicycle was a {bicycles[0].title()}!.' #可以使用.title()对取出的元素进行首字母大写，若是词组则会对所有单词进行首字母大写。
print(message)
```

2. 添加元素：

```python
bicycles = ['trek','cannondale','redline','specialized']
bicycles.append('honda') #用.append()方法在列表末尾添加新的元素。
print(bicycles)

bicycles.insert(1,'yamaha') #用.insert()方法在列表的任何位置插入新元素，此时应指定新元素的索引。
print(bicycles)
```

3. 删除元素：

```python
bicycles = ['trek','cannondale','redline','specialized']
del bicycles[0] #可以使用del函数与列表的索引进行列表元素的删除。
print(bicycles)

popped_bicycle = bicycles.pop(0) #.pop()方法可以删除列表中的任意一个元素，其参数为索引（在无任何参数时删除最后一个元素），并返回它的值作为一个变量。
print(bicycles)
print(popped_bicycle)

bicycles.remove('redline') #.remove()用法可以删除任意一个已知的元素，其参数为元素本身。
print(bicycles)
```

4. 对列表进行永久排序：

```python
cars = ['bmw','audi','toyota','subaru']
cars.sort() #.sort()方法可以按照首字母顺序排列列表中的元素，此方法无法被撤销。另外可以添加 reverse=True 参数使其按字母相反顺序排列。
print(cars)
```

5. 对列表进行临时排序：

```python
cars = ['bmw','audi','toyota','subaru']
print(sorted(cars)) #sort函数可以按首字母顺序返回一个新的列表，且不会影响列表原本的排列。其可以添加 reverse=True 参数。
print(cars)
```

6. 反转列表：

```python
cars = ['bmw','audi','toyota','subaru']
print(cars)

cars.reverse() #.reverse方法可以永久性颠倒列表的排序。
print(cars)
```

7. 查询列表的长度：

```python
cars = ['bmw','audi','toyota','subaru']
len(cars) #len函数直接返回列表长度的数值。
```

# 2. 关于字典(dict)：

+ 一般使用“{}”表示字典。字典是指一系列键值对。每个键都与一个值相关联，可以通过键来访问与之相关联的值。任何对象都可以作为字典中的值。

```python
alien_0 = {'color':'yellow','points':5}
print(alien_0)
```

+ 要访问字典中的值时，可以依次指定字典和放在方括号内的值，或者使用.get()方法：

```python
alien_0 = {'color':'yellow','points':5}
print(alien_0['color'])

try:
    print(alien_0['x_position'])
except KeyError:
    print('The x_position is not defined.') #使用方括号访问值时，若该键未被定义，则会引发Traceback（KeyError）

point_value = alien_0.get('points')
print(point_value)
x_position = alien_0.get('x_position') #使用.get()方法访问值时，即使该键未被定义，也只会返回None，而不是Traceback
print(x_position)
```

+ 要添加键值对时，可以依次指定字典、用方括号括起来的键和相关联的值：

```python
alien_0 = {'color':'yellow','points':5}
print(alien_0)

alien_0['x_position'] = 0
alien_0['y_position'] = 25
print(alien_0)
```

+ 要修改字典中的值时，可以依次指定字典、用方括号括起来的键和与其相关联的新值：

```python
alien_0 = {'color':'yellow','points':5}
print(f'It is {alien_0["color"]} ')

alien_0['color'] = 'blue'
print(f'It is {alien_0["color"]} ')
```

+ 要删除字典中的键值对时，可以用del函数并指定字典和要删除的键：

```python
alien_0 = {'color':'yellow','points':5}
print(alien_0)

del alien_0['color']
print(alien_0)
```

+ 可以将一个较大的字典放在多行中：

```python
favorite_language = {
    'jen': 'python', #此处应缩进
    'sarah': 'c',
    'edward': 'ruby',
    'phil': 'python',
}
print(favorite_language)
```

+ 可以对字典进行遍历：

```python
use_0 = {
    'username':'john',
    'email':'123@123.com',
    'password':'114514',
}

for k,v in use_0.items(): #.items()方法是生成一个视图对象(dict.items)，其内容包含所有键值对。
    print(f'\nKey: {k} \nValue: {v}')
```

+ 可以用.key()方法返回一个由字典中的所有键组成的视图对象（dict_keys），可以理解为打开了一个小窗口访问字典中的内容：

```python
favorite_language = {
    'jen': 'python',
    'sarah': 'c',
    'edward': 'ruby',
    'phil': 'python',
}

for name in favorite_language.keys(): #此处的.keys()可以省略，若直接遍历一个字典，则会默认遍历其所有的键。
    print(name.title())

print('')

for st_name in sorted(favorite_language.keys()): #可以套用sort函数按特定顺序遍历键列表的副本。
    print(st_name.title())
```

+ 可以用.values()方法返回一个由字典中所有值组成的视图对象（dict.values）：

```python
favorite_language = {
    'jen': 'python',
    'sarah': 'c',
    'edward': 'ruby',
    'phil': 'python',
    'linda': 'python'
}

print('The following are the favorite languages:')
for language in favorite_language.values():
    print(language.title())

print('\nThe following are the favorite languages:')
for language_set in set(favorite_language.values()): #可以用set函数创建一个集合，根据集合的唯一性，会自动剔除重复元素。
    print(language_set.title())
```

# 3.关于Lambda匿名函数：

+ 在编写只有一个返回值的比较简单的函数时，可以使用Lambda匿名函数。其牺牲了部分代码可读性，但是具有较高的简洁性。

1. 在求一个二次函数的值时，正常方式如下：

```python
def function(a,b,c,x): #要求提供二次函数中的三个参数和x的取值。
    value = a*x**2 + b*x + c
    return value

print(function(1,2,3,4))
```

2. 由于function函数相当简单，而且只有一个返回值，所以可以对其改写为Lambda匿名函数：

```python
value = lambda a,b,c,x:a*x**2 + b*x + c #编写Lambda函数时，使用关键字lambda，其后跟上数个形参，冒号后为lambda函数返回值。
print(value(1,2,3,4))
```

3. 当然也可以改成下列的写法：

```python
def function(a,b,c):
    return lambda x:a*x**2 + b*x + c

print(function(1,2,3)(4)) #此处的function(1,2,3)应看成f，即print()内部实际为f(4)。
```

# 4.关于Decorator装饰器：

+ Decorator装饰器可以在不改变函数主体的情况下为其封装新功能。

1. 这是一个问候用户的函数：

```python
def greeting():
    print('Hello')

greeting()
```

2. 使用Decorator为其增加几条输出：

```python
def logger(func): #logger()是一个装饰器函数，它可以接受一个函数并返回一个新的函数wrapper()
    def wrapper(): #wrapper()在调用func()前后插入了额外的内容
        print('Start')
        func()
        print('End')
    return wrapper #此处将wrapper()这一函数返回给logger()，相当于logger()接受了一个函数后将其包装一下后输出。

def greeting():
    print('Hello')


greet = logger(greeting) #需要这一步赋值的原因是，第六行中返回的是一个函数，它没有被执行，需要一个变量进行接受并加上()表示执行。当然第六行若是直接返回wrapper()则不需要这一步赋值，因为返回的是被执行的结果。
greet()
```

3. 若Decorator接受的函数需要提供参数，则需要修改写法。

```python
def logger(func):
    def wrapper(*args, **kwargs): #此处*args表示一切位置参数，**kwargs表示一切关键字参数。相当于把func()的所有参数原封不动地赋给wrapper()。
        print('Start')
        value = func(*args, **kwargs)
        print('End')
        return value
    return wrapper #返回的是函数本身，而不是调用它。


def function(a,b,c):
    value_0 = a + b + c
    print('正在计算：')
    return value_0

print(function(1,2,3))
print('')
result = logger(function) #此处将function这一函数放入logger，并且返回wrapper这一函数，它本身被赋给了result这一变量。
print(result(1,2,3))
```

# 5.关于类（class）：

+ 类可以模拟现实中的实物，且可以基于类创建对象。编写类时，可以定义一大类对象都有的通用行为，而类中的对象具有类的一切通用行为，并且可以根据需要赋予每个对象独特的个性：

```python
class Dog:
    def __init__(self, name, age): #初始化dog，具体内容在下一条内解释。
        self.name = name
        self.age = age
        self.color = 'red' #此处为属性提供了默认值。

    def sit(self): #类内定义的函数称为方法。一般使用.xxx()进行调用。
        print(f'{self.name} is now sitting.')

    def roll_over(self):
        print(f'{self.name} is now rolling over.')

    def search_age(self):
        print(f'{self.name} is now {self.age} years old.')

my_dog = Dog('Wangcai',5) #创建一个实例。
my_dog.sit() #调用实例可以使用的方法。
my_dog.roll_over()
my_dog.search_age()
print(my_dog.color)
```

+ 关于__init__( )：

1. 此为一个特殊方法，每当有实例被创建时，它都会自动被运行（即不需要调用.init( )）。
2. \_\_init__( )方法包含的形参中，第一个形参是自指，即指代用户给出的变量名，并且该形参不需要指定。因此在创建实例时，只需要给出除self以外的形参。在上一板块的内容中，self即指代my_dog变量。
3. \_\_init__( )的其他形参被称为属性，根据需要添加。
4. 在\_\_init__( )的下一行，是将变量的属性与给出的值相关联，以便通过“变量名.属性”访问之。

+ 修改属性的值：

```python
class Car:
    def __init__(self,make,model,year):
        self.make = make
        self.model = model
        self.year = year
        self.odometer_reading = 0

    def read_odometer(self):
        print(f'This car has {self.odometer_reading} miles on it.')

    def update_odometer(self,mileage):
        self.odometer_reading = mileage


my_new_car = Car('Audi','A4',2016)
my_new_car.read_odometer()

my_new_car.odometer_reading = 23 #可以直接通过实例修改它的属性。
my_new_car.read_odometer()

my_new_car.update_odometer(26) #可以通过类中的方法修改属性的值。
my_new_car.read_odometer()
```

+ 如果要编写的类是另一个现有的类的特殊版本，可以使用继承。当一个类继承另一个类时，将会自动获得另一个类的所有属性和方法。原有的类称为父类，新生成的类称为子类：

```python
class Car:
    def __init__(self,make,model,year):
        self.make = make
        self.model = model
        self.year = year
        self.odometer_reading = 0

    def read_odometer(self):
        print(f'This car has {self.odometer_reading} miles on it.')

    def get_descriptive_name(self):
        long_name = str(self.year) + ' ' + self.make + ' ' + self.model
        return long_name.title()

    def update_odometer(self,mileage):
        self.odometer_reading = mileage

class ElectricCar(Car):
    def __init__(self,make,model,year): #此处的逻辑是，先自动初始化输入的实例，并且使用super函数，调用Car类中的方法__init__()，让ElectricCar实例包含这个方法中定义的所有属性。
        super().__init__(make,model,year) #相当于self.xxx = xxx

my_tesla = ElectricCar('tesla','model s',2019)
print(my_tesla.get_descriptive_name())
```

+ 可以修改子类的属性与方法：

```python
class Car:
    def __init__(self,make,model,year):
        self.make = make
        self.model = model
        self.year = year
        self.odometer_reading = 0
        self.gas = 10

    def read_odometer(self):
        print(f'This car has {self.odometer_reading} miles on it.')

    def get_descriptive_name(self):
        long_name = str(self.year) + ' ' + self.make + ' ' + self.model
        return long_name.title()

    def update_odometer(self,mileage):
        self.odometer_reading = mileage

    def fill_gas_tank(self):
        self.gas += 10
        print(self.gas)

class ElectricCar(Car):
    def __init__(self,make,model,year):
        super().__init__(make,model,year)
        self.battery_size = 75 #此处直接定义了一个电车特有的属性。

    def describe_battery(self):
        print(f'The battery size is: {self.battery_size}')

    def fill_gas_tank(self):
        print(f'This car does not have a gas tank.') #此处改写了父类中的fill_gas_tank方法，由于电车没有油箱，故只返回一句提示。

my_tesla = ElectricCar('tesla','model s',2019)
my_tesla.describe_battery()
my_tesla.fill_gas_tank()
```

+ 可以将实例用作属性：

```python
class Car:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year
        self.odometer_reading = 0
        self.gas = 10

    def read_odometer(self):
        print(f'This car has {self.odometer_reading} miles on it.')

    def get_descriptive_name(self):
        long_name = str(self.year) + ' ' + self.make + ' ' + self.model
        return long_name.title()

    def update_odometer(self, mileage):
        self.odometer_reading = mileage

    def fill_gas_tank(self):
        self.gas += 10
        print(self.gas)

class Battery:
    def __init__(self, battery_size = 75):
        self.battery_size = battery_size

    def describe_battery(self):
        print(f'The battery size is: {self.battery_size}')

class ElectricCar(Car):
    def __init__(self, make, model, year):
        super().__init__(make, model, year)
        self.battery_size = Battery()  #此处创建了一个battery实例。

    def describe_battery(self):
        print(f'The battery size is: {self.battery_size}')

    def fill_gas_tank(self):
        print(f'This car does not have a gas tank.')


my_tesla = ElectricCar('tesla', 'model s', 2019)
my_tesla.battery_size.describe_battery()
```

# 6.关于魔法方法（Magic Methods）：

+ 魔法方法是指python中以两个下划线开头和结尾的特殊方法。它可以让类的实例在特定情况下自动触发特定行为。实际上，魔法方法是语法糖的底层机制：

```python
class Student:
    def __init__(self,name,age):
        self.name = name
        self.age = age

    def __str__(self): #此处的__str__()将在print这一语法糖被执行时自动执行，即调用print时，让这个实例自动返回被定义的内容。可以理解为实例的自述。
        return f'Student {self.name} is {self.age} years old.'

    def __len__(self): #此处的__len__()将在len这一语法糖被执行时自动执行，当然返回值不一定是名字的长度，它可以是任何东西的长度，只是在调用len时，会自动把该长度打印出来。
        l = len(self.name)
        return l

s = Student('James',25)
print(s)
len(s)
```

+ 值得注意的是，魔法方法一般都对返回值有所要求，如上述的__str__( )，它的返回值要求是一个字符串，否则会引发Traceback，而__len__( )要求返回的是一个整数。

# 7.关于re正则表达式：

+ re正则表达式是一个模块，它一般被用来通过一小段规则字符串去匹配大量复杂文本，类似于“查找”功能，但是比“查找”更权威，更灵活，而且一次输入即得所有结果。

+ re模块提供了一系列方法：

1. re.match( ) >>> 从字符串的开头开始匹配。
2. re.search( ) >>> 在整个字符串中查找第一个匹配的内容。
3. re.findall( ) >>> 返回所有匹配项的列表。
4. re.sub( ) >>> 用于替换匹配到的内容。
5. re.split( ) >>> 用于按照匹配规则分割字符串。
6. re.compile( ) >>> 预编译一个正则表达式。

+ 核心用法（摘录于ChatGPT）

| 符号      | 含义              | 示例        | 匹配结果                  |      |                  |
| ------- | --------------- | --------- | --------------------- | ---- | ---------------- |
| `.`     | 任意一个字符（除换行）     | `a.c`     | 匹配 `abc`, `axc`       |      |                  |
| `\d`    | 数字（0-9）         | `\d\d\d`  | 匹配三位数字                |      |                  |
| `\w`    | 字母、数字、下划线       | `\w+`     | 匹配单词                  |      |                  |
| `\s`    | 空格、制表符等         | `\s+`     | 匹配空白部分                |      |                  |
| `+`     | 前一个字符重复一次或多次    | `a+`      | 匹配 “a”, “aa”, “aaa”   |      |                  |
| `*`     | 前一个字符重复零次或多次    | `ba*`     | 匹配 “b”, “ba”, “baa”   |      |                  |
| `?`     | 前一个字符出现 0 或 1 次 | `colou?r` | 匹配 “color” 或 “colour” |      |                  |
| `[]`    | 字符集（任选一个）       | `[abc]`   | 匹配 “a”, “b”, “c”      |      |                  |
| `[^]`   | 非字符集            | `[^0-9]`  | 匹配非数字                 |      |                  |
| `{m,n}` | 重复 m 到 n 次      | `\d{2,4}` | 匹配 2~4 位数字            |      |                  |
| `^`     | 行开头             | `^Hello`  | 匹配以 Hello 开头的字符串      |      |                  |

# 8.关于列表推导式：

+ 列表推导式是基于可迭代对象快速创建新的列表的方法。可以将其理解为，在创建一个列表时，将for循环与list.append()合并为一行的写法。这种写法既具有简洁性又具有可读性：

```python
l1 = []

for i in range(1,10): #正常创建一个列表的过程，此处可迭代对象指的是range(1,10)。
    l1.append(i**2)

print(f'l1 = {l1}')

l2 = [i**2 for i in range(1,10)] #此处的l2即是使用列表推导式创建的，其基本逻辑是，先声明需要加入列表的变量，然后再使用for循环产生需要的变量，python会自动把for产生的变量通过变换后加入列表。
print(f'l2 = {l2}')
```

+ 当然，可以对循环进行筛选或嵌套：

```python
l1 = [i**2 for i in range(1,10) if i % 2 == 0] #此处筛选出偶数并对其进行取平方操作。
print(f'l1 = {l1}')

l2 = [(x,y) for x in range(1,4) for y in range(1,4)] #先声明两个变量，再分别使用for产生之
print(f'l2 = {l2}')
```

# 9.关于类型注释（Type Hint）：

+ 类型注释一般被用于在变量、函数参数和函数返回值上添加关于数据类型的注解。它可以提高维护代码的效率，并使得错误被尽早发现。类型注释是可选的，它不会改变python的运行方式。

1. 当为变量添加注释时，可以使用冒号：

```python
name: str = 'Alice' #在冒号后加上应该输入的变量类型。
age: int = 30
active: bool = True

print(name,age,active)
```

2. 当为函数的参数和返回值注释时，可以分别使用冒号和箭头：

```python
def calculate(length: float, width: float) -> float: #此处定义的函数，使用类型注释标明了输入和输出值都应为浮点数。
    return length * width

print(calculate(10.5,2.0))
```

3. 当为复杂类型注释时，应使用typing模块中的特定类型，具体类型见下（摘录自ChatGPT）：

| 类型 | 描述 | 示例用法 |
|:----:|:-----|:----------|
| `List` | 列表（元素类型需要指定） | `data: List[int] = [1, 2, 3]` |
| `Dict` | 字典（键和值类型都需要指定） | `settings: Dict[str, str] = {"key": "value"}` |
| `Tuple` | 元组（元素类型和顺序需要指定） | `point: Tuple[int, int] = (10, 20)` |
| `Optional` | 可选类型，表示可以接受某种类型的值，或者 `None`（空）。<br>*(Python 3.10+ 可用 `int &#124; None`)* | `age: Optional[int] = None` |
| `Union` | 联合类型，表示可以是多种类型中的任意一种。<br>*(Python 3.10+ 可用 `int &#124; str`)* | `value: Union[int, str] = 42` |




```python
from typing import Dict, Optional, List #引入typing模块中的注释类型。

dic = {'1':'linda', '2':'alan', '3':'alex'}

def receive_names(names: Dict[str, str]) -> List[str]: #此处期望接受一个全是字符串的字典，并且返回一个包含所有用户名字的列表。
    return list(names.values())

print(receive_names(dic))

def get_user(user_index: int) -> Optional[str]: #在字典中根据索引查找用户，找到了就返回他的名字（字符串），没找到就返回None
    return dic.get(f'{user_index}')


print('')
print(get_user(2))
print('')
print(get_user(4))
```

# 10.关于Generator生成器：

+ 生成器实际上是一种迭代器，它使用yield关键字来产出需要的数据，具有按需生成、内存友好的特点。

+ 定义并访问一个生成器：

```python
def gen_even():
    number = 0
    yield number#此处用yield关键字记录下当前的number，并将它传递给gen这一变量
    print(number)
    number += 2
    yield number
    print(number)
    number += 2
    yield number
    print(number)

print('begin')
gen = gen_even() #此处调用gen_even，但是其不会打印出各步所得的数字，而是返回一个可以迭代的生成器。
print('end')

l = []

for i in gen: #在执行此遍历时，才会开始执行gen_even的内容，其中的print函数被执行并打印出每一步所得到的数字，并且变量i会被赋值为yield的内容。每次循环gen_even函数都会停在yield一行，故其节约资源。
    l.append(i)

print(l)
```

+ 在处理无限序列时，可以用next函数获取生成器中的下一个值，这也是生成器的优势所在：随用随取。

```python
def gen_even():
    number = 0
    while True:
        yield number
        number += 2

even_gen = gen_even()

print(next(even_gen))
print(next(even_gen))
print(next(even_gen))
```

+ 与列表推导式类似，生成器也具有生成器表达式，它使用的是圆括号：

```python
l = [x**2 for x in range(1,10)]
print(l)

print('')

g = (x**2 for x in range(1,10)) #此处相当于每生成一个x**2就yield一次，并且g变量储存的是生成器对象，它在被迭代时按需生成值。。
print(g) #这一行的运行结果表明g是一个生成器。

result = []
for i in g:  #即访问生成器中内容的方法，当然也可以用next函数。
    result.append(i)
print(result)


```
