# 1.关于列表(list)：


+ 一般使用"[ ]"表示列表。列表可用于储存一组有特定排列方式的数据，可以将任何东西加入列表，且列表中的元素之间没有任何关系。列表可以被遍历。

```python
bicycles = ['trek','cannondale','redline','specialized']
print(bicycles)
```

+ 访问列表中的元素时，可以使用列表名加索引的形式，列表的第一个元素的索引是0。列表的索引也可以反着看，但是最后一个元素的索引是-1。

```python
print(bicycles[0])
```

+ 在列表中修改元素时，可以直接使用列表名加索引，并令其为一个新的值。

```python
bicycles = ['trek','cannondale','redline','specialized']
bicycles[0] = 'ducati'
print(bicycles)
```

+ 列表的用法

1. 首字母大写：

```python
bicycles = ['trek','cannondale','redline','specialized']
message = f'My first bicycle was a {bicycles[0].title()}!.' #可以使用.title()对取出的元素进行首字母大写，若是词组则会对所有单词进行首字母大写。
print(message)
```

2. 添加元素：

```python
bicycles = ['trek','cannondale','redline','specialized']
bicycles.append('honda') #用.append()方法在列表末尾添加新的元素。
print(bicycles)

bicycles.insert(1,'yamaha') #用.insert()方法在列表的任何位置插入新元素，此时应指定新元素的索引。
print(bicycles)
```

3. 删除元素：

```python
bicycles = ['trek','cannondale','redline','specialized']
del bicycles[0] #可以使用del函数与列表的索引进行列表元素的删除。
print(bicycles)

popped_bicycle = bicycles.pop(0) #.pop()方法可以删除列表中的任意一个元素，其参数为索引（在无任何参数时删除最后一个元素），并返回它的值作为一个变量。
print(bicycles)
print(popped_bicycle)

bicycles.remove('redline') #.remove()用法可以删除任意一个已知的元素，其参数为元素本身。
print(bicycles)
```

4. 对列表进行永久排序：

```python
cars = ['bmw','audi','toyota','subaru']
cars.sort() #.sort()方法可以按照首字母顺序排列列表中的元素，此方法无法被撤销。另外可以添加 reverse=True 参数使其按字母相反顺序排列。
print(cars)
```

5. 对列表进行临时排序：

```python
cars = ['bmw','audi','toyota','subaru']
print(sorted(cars)) #sort()函数可以按首字母顺序输出列表，且不会影响列表原本的排列。其可以添加 reverse=True 参数。
print(cars)
```

6. 反转列表：

```python
cars = ['bmw','audi','toyota','subaru']
print(cars)

cars.reverse() #.reverse方法可以永久性颠倒列表的排序。
print(cars)
```

7. 查询列表的长度：

```python
cars = ['bmw','audi','toyota','subaru']
len(cars) #len()函数直接返回列表长度的数值。
```
