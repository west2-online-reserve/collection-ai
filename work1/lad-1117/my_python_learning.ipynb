# 1.关于列表(list)：


+ 一般使用"[ ]"表示列表。列表可用于储存一组有特定排列方式的数据，可以将任何东西加入列表，且列表中的元素之间没有任何关系。列表可以被遍历。

```python
bicycles = ['trek','cannondale','redline','specialized']
print(bicycles)
```

+ 访问列表中的元素时，可以使用列表名加索引的形式，列表的第一个元素的索引是0。列表的索引也可以反着看，但是最后一个元素的索引是-1。

```python
print(bicycles[0])
```

+ 在列表中修改元素时，可以直接使用列表名加索引，并令其为一个新的值。

```python
bicycles = ['trek','cannondale','redline','specialized']
bicycles[0] = 'ducati'
print(bicycles)
```

+ 列表的用法

1. 首字母大写：

```python
bicycles = ['trek','cannondale','redline','specialized']
message = f'My first bicycle was a {bicycles[0].title()}!.' #可以使用.title()对取出的元素进行首字母大写，若是词组则会对所有单词进行首字母大写。
print(message)
```

2. 添加元素：

```python
bicycles = ['trek','cannondale','redline','specialized']
bicycles.append('honda') #用.append()方法在列表末尾添加新的元素。
print(bicycles)

bicycles.insert(1,'yamaha') #用.insert()方法在列表的任何位置插入新元素，此时应指定新元素的索引。
print(bicycles)
```

3. 删除元素：

```python
bicycles = ['trek','cannondale','redline','specialized']
del bicycles[0] #可以使用del函数与列表的索引进行列表元素的删除。
print(bicycles)

popped_bicycle = bicycles.pop(0) #.pop()方法可以删除列表中的任意一个元素，其参数为索引（在无任何参数时删除最后一个元素），并返回它的值作为一个变量。
print(bicycles)
print(popped_bicycle)

bicycles.remove('redline') #.remove()用法可以删除任意一个已知的元素，其参数为元素本身。
print(bicycles)
```

4. 对列表进行永久排序：

```python
cars = ['bmw','audi','toyota','subaru']
cars.sort() #.sort()方法可以按照首字母顺序排列列表中的元素，此方法无法被撤销。另外可以添加 reverse=True 参数使其按字母相反顺序排列。
print(cars)
```

5. 对列表进行临时排序：

```python
cars = ['bmw','audi','toyota','subaru']
print(sorted(cars)) #sort()函数可以按首字母顺序输出列表，且不会影响列表原本的排列。其可以添加 reverse=True 参数。
print(cars)
```

6. 反转列表：

```python
cars = ['bmw','audi','toyota','subaru']
print(cars)

cars.reverse() #.reverse方法可以永久性颠倒列表的排序。
print(cars)
```

7. 查询列表的长度：

```python
cars = ['bmw','audi','toyota','subaru']
len(cars) #len()函数直接返回列表长度的数值。
```

# 2. 关于字典(dict)：

+ 一般使用“{}”表示字典。字典是指一系列键值对。每个键都与一个值相关联，可以通过键来访问与之相关联的值。任何对象都可以作为字典中的值。

```python
alien_0 = {'color':'yellow','points':5}
print(alien_0)
```

+ 要访问字典中的值时，可以依次指定字典和放在方括号内的值，或者使用.get()方法。

```python
alien_0 = {'color':'yellow','points':5}
print(alien_0['color'])

try:
    print(alien_0['x_position'])
except KeyError:
    print('The x_position is not defined.') #使用方括号访问值时，若该键未被定义，则会引发Traceback

point_value = alien_0.get('points')
print(point_value)
x_position = alien_0.get('x_position') #使用.get()方法访问值时，即使该键未被定义，也只会返回一个空值，而不是Traceback
print(x_position)
```

+ 要添加键值对时，可以依次指定字典、用方括号括起来的键和相关联的值。

```python
alien_0 = {'color':'yellow','points':5}
print(alien_0)

alien_0['x_position'] = 0
alien_0['y_position'] = 25
print(alien_0)
```

+ 要修改字典中的值时，可以依次指定字典、用方括号括起来的键和与其相关联的新值。

```python
alien_0 = {'color':'yellow','points':5}
print(f'It is {alien_0["color"]} ')

alien_0['color'] = 'blue'
print(f'It is {alien_0["color"]} ')
```

+ 要删除字典中的键值对时，可以用del()函数并指定字典和要删除的键。

```python
alien_0 = {'color':'yellow','points':5}
print(alien_0)

del alien_0['color']
print(alien_0)
```

+ 可以将一个较大的字典放在多行中。

```python
favorite_language = {
    'jen': 'python', #此处应缩进
    'sarah': 'c',
    'edward': 'ruby',
    'phil': 'python',
}
print(favorite_language)
```

+ 可以对字典进行遍历。

```python
use_0 = {
    'username':'john',
    'email':'123@123.com',
    'password':'114514',
}

for k,v in use_0.items(): #。items()方法是取出所有键值对存入一个列表，并返回这个列表，所以此步骤相当于把字典转换为列表再遍历列表。
    print(f'\nKey: {k} \nValue: {v}')
```

+ 可以用.key()方法返回一个由字典中的所有键组成的列表。

```python
favorite_language = {
    'jen': 'python',
    'sarah': 'c',
    'edward': 'ruby',
    'phil': 'python',
}

for name in favorite_language.keys(): #此处的.keys()可以省略，若直接遍历一个字典，则会默认遍历其所有的键。
    print(name.title())

print('')

for st_name in sorted(favorite_language.keys()): #可以套用sort()函数按特定顺序遍历键列表的副本。
    print(st_name.title())
```

+ 可以用.values()方法返回一个由字典中所有值组成的列表。

```python
favorite_language = {
    'jen': 'python',
    'sarah': 'c',
    'edward': 'ruby',
    'phil': 'python',
    'linda': 'python'
}

print('The following are the favorite languages:')
for language in favorite_language.values():
    print(language.title())

print('\nThe following are the favorite languages:')
for language_set in set(favorite_language.values()): #可以用set()函数创建一个集合，根据集合的唯一性，会自动剔除重复元素。
    print(language_set.title())
```

# 3.关于Lambda匿名函数：

+ 在编写只有一个返回值的比较简单的函数时，可以使用Lambda匿名函数。其牺牲了部分代码可读性，但是具有较高的简洁性。

1. 在求一个二次函数的值时，正常方式如下：

```python
def function(a,b,c,x): #要求提供二次函数中的三个参数和x的取值。
    value = a*x**2 + b*x + c
    return value

print(function(1,2,3,4))
```

2. 由于function()函数相当简单，而且只有一个返回值，所以可以对其改写为Lambda匿名函数。

```python
value = lambda a,b,c,x:a*x**2 + b*x + c #编写Lambda函数时，使用函数名lambda，其后跟上数个形参，冒号后为lambda函数返回值。
print(value(1,2,3,4))
```

3. 当然也可以改成下列的写法。

```python
def function(a,b,c):
    return lambda x:a*x**2 + b*x + c

print(function(1,2,3)(4)) #此处的function(1,2,3)应看成f，即print()内部实际为f(4)。
```

# 4.关于Decorator装饰器：

+ Decorator装饰器可以在不改变函数主体的情况下为其临时动态地增加新功能。

1. 这是一个问候用户的函数。

```python
def greeting():
    print('Hello')

greeting()
```

2. 使用Decorator为其增加几条输出。

```python
def logger(func): #logger()是一个装饰器函数，它可以接受一个函数并返回一个新的函数wrapper()
    def wrapper(): #wrapper()在调用func()前后插入了额外的内容
        print('Start')
        func()
        print('End')
    return wrapper #此处将wrapper()这一函数返回给logger()，相当于logger()接受了一个函数后将其包装一下后输出。

def greeting():
    print('Hello')


greet = logger(greeting) #需要这一步赋值的原因是，第六行中返回的是一个函数，它没有被执行，需要一个变量进行接受并加上()表示执行。当然第六行若是直接返回wrapper()则不需要这一步赋值，因为返回的是被执行的结果。
greet()
```

3. 若Decorator接受的函数需要提供参数，则需要修改写法。

```python
def logger(func):
    def wrapper(*args, **kwargs): #此处*args表示一切位置参数，**kwargs表示一切关键字参数。相当于把func()的所有参数原封不动地赋给wrapper()。
        print('Start')
        value = func(*args, **kwargs)
        print('End')
        return value
    return wrapper #返回的是函数本身，而不是调用它。


def function(a,b,c):
    value_0 = a + b + c
    print('正在计算：')
    return value_0

print(function(1,2,3))
print('')
result = logger(function)
print(result(1,2,3))
```
