# 函数

## Lamda
- 非显式定义的函数

```python
list(map(lambda x : x+1,[1,2,3]))
```

## Decorator
- 动态增加函数功能的函数

```python
def log(func):
    def wrapper(*args,**kw):
        print("正在调用：",func.__name__)
        return func(*args,**kw)
    return wrapper

@log
def f():
    print(1)

f()
```

# 面向对象
## class
- 具有特定功能的数据与方法的抽象
## 魔法方法
- 改变类的内在属性和遇到特定方法时的表现
## OOP
- 以类的思想对需要实现的数据与方法进行抽象进而进行编程

```python
class Cls:
    def __init__(self,data):
        self.data=data
    
    def __str__(self):
        return self.data
    
cls=Cls("data")
print(cls)
```

# 文本处理
## re正则表达式
- 文本匹配的一种规则

```python
import re
str_="username: message 2025.6.4"
m=re.match(r'([a-z]+):[\s]([0-9a-zA-Z\s+])[\s]([0-9]+).([0-9]+).([0-9])',str_)
```
