# 矩阵操作与数据分析

```python
import numpy as np
```

## 数组创建

```python
data_matrix = np.arange(0, 100).reshape(10, 10)
data_matrix
```

## 子矩阵提取

```python
data_matrix[3:7, 3:7]
```

## 条件筛选与赋值

```python
mask = data_matrix > 75
data_matrix[mask] = 0

data_matrix
```

## 向量化运算

有点尴尬，发现必须用`astype`转一下类型，`*=`之类的也是`inplace`操作，比`np.multiply`糖一点

```python
data_matrix = data_matrix.astype(np.float64)
data_matrix *= 0.8

data_matrix
```

## 数据聚合与定位

```python
max_value = data_matrix.max()
index = np.unravel_index(data_matrix.argmax(), data_matrix.shape)

print(max_value, index)
```
