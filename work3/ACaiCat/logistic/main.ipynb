# Logistic 函数

```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.sans-serif'] = ['SimHei']  # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False    # 用来正常显示负号
```

$f(x) = \frac{L}{1 + e^{-k(x-x_0)}}$

```python
# noinspection InconsistentLineSeparators
def logistic(_x: np.ndarray, l: float, k: float, x0: float) -> np.ndarray:
    _y = l / (np.e ** ((-k) * (_x - x0)) + 1)
    return _y


x = np.linspace(-10, 10, 1000)
y = logistic(x, 2, 1, 1)

fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
```

### $L$
上渐近线$y=L$，纵向拉伸图像
### $k$
陡度，$k$越大图像越陡峭，$k$越小图像越平缓
### $x_0$
图像中心$x$坐标，改变可以水平移动图像

```python
x = np.linspace(-10, 10, 1000)

y1 = logistic(x, 2, 1, 1)
y2 = logistic(x, 1, 1, 1)
y3 = logistic(x, 2, 8, 1)
y4 = logistic(x, 2, 1, 4)

fig, axs = plt.subplots(2, 2, layout='constrained')

axs[0, 0].plot(x, y1, color='C1', label="对照")
axs[0, 0].plot(x, y2, color='C2', label="$L=1$")
axs[0, 0].plot(x, y3, color='C3', label="$k=8$")
axs[0, 0].plot(x, y4, color='C4', label="$x_0=4$")
axs[0, 0].legend()
axs[0, 0].set_title("对照")

axs[0, 1].plot(x, y1, color='C1', label="对照")
axs[0, 1].plot(x, y2, color='C2', label="$L=1$")
axs[0, 1].plot(x, np.full_like(x, 2), linestyle='--', color='gray')
axs[0, 1].plot(x, np.full_like(x, 1), linestyle='--', color='gray')
axs[0, 1].annotate('渐近线更低', xy=(2, 1), xytext=(-10, 0.8), color='orange')
axs[0, 1].legend()
axs[0, 1].set_title("$L=1$")

axs[1, 0].plot(x, y1, color='C1', label="对照")
axs[1, 0].plot(x, y3, color='C3', label="$k=8$")
axs[1, 0].annotate('更加陡峭', xy=(0, 1), xytext=(-10, 0.9), color='orange',
                   arrowprops=dict(arrowstyle='->'))
axs[1, 0].legend()
axs[1, 0].set_title("$k=8$")

axs[1, 1].plot(x, y1, color='C1', label="对照")
axs[1, 1].plot(x, y4, color='C4', label="$x_0=4$")
axs[1, 1].plot(x, np.full_like(x, 1), linestyle='--', color='gray')
axs[1, 1].annotate('整体向右平移', xy=(4, 1), xytext=(4.4, 0.3), color='orange',
                   arrowprops=dict(arrowstyle='->'))
axs[1, 1].legend()
axs[1, 1].set_title("$x_0=4$")

plt.show()
```
