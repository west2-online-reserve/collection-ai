# 广播机制与向量化距离计算

```python
import numpy as np
```

## 数据生成

```python
rng = np.random.default_rng()
points_A = rng.integers(0, 100, size=(5, 2), endpoint=True)
points_B = rng.integers(0, 100, size=(8, 2), endpoint=True)
```

## 距离矩阵计算

```python
# def euclidean_distance(q, p) -> int:

# A -> (5, 1, 2)
# B -> (1, 8, 2)
# O =  (5, 8, 2)

distance_matrix: np.ndarray = np.sqrt(np.sum((points_A[:, np.newaxis, :] - points_B[np.newaxis, :, :]) ** 2, axis=2))

#

distance_matrix
```

## 轴向数据聚合

```python
min_distance = distance_matrix.min(axis=1)

min_distance
```

## 复合条件查询

```python
indexes_mark = np.any(distance_matrix<20, axis=0)
indexes = np.arange(8)[indexes_mark]

indexes
```
