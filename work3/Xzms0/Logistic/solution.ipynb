### 作业二：Logistic 函数（Matplotlib）

#### 背景知识

生物必修二里，有一张关于种群数量分析的「S」形曲线图，那条美丽的曲线就是大名鼎鼎的 **Logistic 函数**，它的数学表达式是：

$f(x) = \frac{L}{1 + e^{-k(x-x_0)}}$

其中：

- `L` 代表曲线的饱和值（最大值）。
- `k` 代表曲线的陡峭程度。
- `x0` 代表 S 形曲线中点的位置。

#### 要求

1. 请使用 `matplotlib` 绘制一个标准的 Logistic 函数图像，其中参数设置为：`L=1`、`k=1`、`x0=0`。
2. 为了更好地理解，请尝试在同一张图或多张子图中，绘制不同 `L`、`k`、`x0` 参数下的 Logistic 函数，并用图例或文字简单说明，改变每个参数是如何影响函数形状的。

> 其实这个函数在深度学习里经常被使用到。

#### 分析

+ 首先定义 Logistic 函数

```python
from math import e

import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

def logistic(L = 1, k = 1, x0 = 0):
    x = np.arange(-10, 10, 0.1, dtype=float)
    y =  L / (1 + e ** (-k * (x - x0)))

    plt.plot(x, y, label = f"L={L}, k={k}, x0={x0}")
    plt.plot(x0, L / 2, marker = 'o')
    plt.text(x0, L / 2, f"({x0}, {L / 2})", ha = "left", va = "top")
    plt.legend()
```

+ 绘制一条标准 Logistic 函数图像

```python
L, k, x0 = 1, 1, 0
logistic(L = L, k = k, x0 = x0)
```

+ 改变参数`L`，绘制第二条 Logistic 函数图像。

```python
L, k, x0 = 1, 1, 0
logistic(L = L, k = k, x0 = x0)

L, k, x0 = 2, 1, 0
logistic(L = L, k = k, x0 = x0)
```

+ **分析** 改变了参数`L`后，函数的Y轴最大值也发生了改变，而中心点的X坐标确没有改变。

+ 改变参数`k`，再次绘制一条新的 Logistic 函数图像。

```python
L, k, x0 = 1, 1, 0
logistic(L = L, k = k, x0 = x0)

L, k, x0 = 2, 1, 0
logistic(L = L, k = k, x0 = x0)

L, k, x0 = 1, 2, 0
logistic(L = L, k = k, x0 = x0)
```

+ **分析** 参数`k`不影响函数的最大值和中心点位置，只影响函数曲线的陡峭程度。随着`k`变大，函数曲线越陡峭。

+ 改变参数`x0`，再次绘制新的 Logistic 函数图像。

```python
L, k, x0 = 1, 1, 0
logistic(L = L, k = k, x0 = x0)

L, k, x0 = 2, 1, 0
logistic(L = L, k = k, x0 = x0)

L, k, x0 = 1, 2, 0
logistic(L = L, k = k, x0 = x0)

L, k, x0 = 1, 1, -3
logistic(L = L, k = k, x0 = x0)

L, k, x0 = 2, 1, -4
logistic(L = L, k = k, x0 = x0)
```

+ **分析** 改变参数`x0`不会影响函数的最大值以及函数的变化率，只是让函数图像整体在X轴上移动。
