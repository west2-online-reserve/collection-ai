```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
def logistic(x,l,k,x0):
    return l/(1+np.exp(-k*(x-x0)))
x=np.linspace(-10,10,100)
y=logistic(x,1,1,0)
plt.plot(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
x=np.linspace(-10,10,100)

plt.subplot(3,2,1)
y=logistic(x,1,1,0)
plt.plot(x,y)
plt.title("L=1,k=1,x0=0")

plt.subplot(3,2,2)
y=logistic(x,2,1,0)
plt.plot(x,y)
plt.title("L=2,k=1,x0=0")

plt.subplot(3,2,3)
y=logistic(x,1,1,0)
plt.plot(x,y)
plt.title("L=1,k=1,x0=0")

plt.subplot(3,2,4)
y=logistic(x,1,2,0)
plt.plot(x,y)
plt.title("L=1,k=2,x0=0")
plt.subplot(3,2,5)
y=logistic(x,1,1,0)
plt.plot(x,y)
plt.title("L=1,k=1,x0=0")

plt.subplot(3,2,6)
y=logistic(x,1,1,5)
plt.plot(x,y)
plt.title("L=1_k=1_x0=5")

plt.tight_layout()
plt.show()
```
