```python
import matplotlib.pyplot as plt

# 数据
frequency = [10, 100, 1000, 10000, 100000]  # 频率 (Hz)
equivalent_resistance = [41.67, 41.67, 86.96, 219.51, 4306.12]  # 等效电阻 (Ω)

# 绘制图表
plt.figure(figsize=(8, 2.5))
plt.plot(frequency, equivalent_resistance, marker='o', label="Equivalent Resistance")

# 设置对数坐标轴
plt.xscale('log')

# 添加标题和标签
plt.title("Equivalent Resistance vs Frequency", fontsize=14)
plt.xlabel("Frequency (Hz) - Log Scale", fontsize=12)
plt.ylabel("Equivalent Resistance (Ω)", fontsize=12)
plt.legend()
plt.grid(True, which="both", linestyle="--", linewidth=0.5)

#添加水印
plt.text(0.5, 0.9, 'By ZKS AND WJH', fontsize=20, color='gray', ha='center', va='center', alpha=0.3, transform=plt.gcf().transFigure)


# 显示图表
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the chart
years = [2023, 2024, 2029]
north_america = [18, 20, 40]
europe = [8, 10, 20]
asia_pacific = [5, 7, 15]
rest_of_world = [2, 3, 6]

# Calculate the cumulative values for stacking
cumulative_europe = np.array(north_america) + np.array(europe)
cumulative_asia_pacific = cumulative_europe + np.array(asia_pacific)
cumulative_rest_of_world = cumulative_asia_pacific + np.array(rest_of_world)

# Set up the figure and axis with a dark theme
plt.figure(figsize=(12, 6), facecolor="#0D1B2A")  # Background color for the figure
ax = plt.gca()
ax.set_facecolor("#0D1B2A")  # Background color for the axis

# Plot the stacked bar chart with adapted colors
plt.bar(
    years, north_america, color="royalblue", edgecolor="black", label="North America"
)
plt.bar(
    years,
    europe,
    bottom=north_america,
    color="mediumpurple",
    edgecolor="black",
    label="Europe",
)
plt.bar(
    years,
    asia_pacific,
    bottom=cumulative_europe,
    color="lightseagreen",
    edgecolor="black",
    label="Asia Pacific",
)
plt.bar(
    years,
    rest_of_world,
    bottom=cumulative_asia_pacific,
    color="greenyellow",
    edgecolor="black",
    label="Rest of the World (RoW)",
)

# Add CAGR text
plt.text(2023.7, 70, "CAGR of\n34.2%", fontsize=35, color="#00FFFF", weight="bold")

# Highlight the 2029 value and draw an arrow
plt.text(
    2029,
    96.3,
    "96.3",
    fontsize=10,
    weight="bold",
    va="bottom",
    ha="center",
    color="white",
)
plt.annotate(
    "",
    xy=(2029, 96.3),
    xytext=(2024.5, 40),
    arrowprops=dict(arrowstyle="->", color="#00FFFF", lw=1.5),
)

# Add title and labels with light-colored fonts
plt.title(
    "Augmented and Virtual Reality Market\nGlobal Forecast to 2029 (USD BN)",
    fontsize=16,
    weight="bold",
    color="#00FFFF",
)
plt.xlabel("Year", fontsize=12, color="white")
plt.ylabel("Market Value (USD BN)", fontsize=12, color="white")
plt.xticks(years, fontsize=10, color="white")
plt.yticks(range(0, 101, 20), fontsize=10, color="white")
plt.ylim(0, 100)

# Add legend with light-colored fonts
plt.legend(
    title="Regions",
    fontsize=10,
    loc="upper left",
    title_fontsize=10,
    facecolor="#0D1B2A",
    edgecolor="white",
    labelcolor="white",
)

# Add gridlines for better readability with lighter lines
plt.grid(axis="y", linestyle="--", alpha=0.4, color="white")

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


# 定义 Logistic 函数
def logistic(x, L, k, x0):
    return L / (1 + np.exp(-k * (x - x0)))


fig, ax = plt.subplots(figsize=(8, 5))

x = np.linspace(-10, 10, 500)

# 字典，储存参数
params = [
    {"L": 1, "k": 1, "x0": 0},
    {"L": 2, "k": 1, "x0": 0},
    {"L": 1, "k": 0.5, "x0": 0},
    {"L": 1, "k": 1.5, "x0": 0},
    {"L": 1, "k": 1, "x0": -2},
    {"L": 1, "k": 1, "x0": 2},
]

# 为每组参数绘制 Logistic 函数
for param in params:
    y = logistic(x, param["L"], param["k"], param["x0"])
    ax.plot(x, y, label=f"L={param['L']}, k={param['k']}, x0={param['x0']}")


ax.legend(fontsize=10)
ax.set_title("Logistic Functions with Different Parameters", fontsize=14)
ax.set_xlabel("x", fontsize=12)
ax.set_ylabel("f(x)", fontsize=12)
ax.grid(alpha=0.3)

plt.show()
```
