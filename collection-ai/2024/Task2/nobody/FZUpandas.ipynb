```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv(r'C:\Users\宋明哲\PycharmProjects\task2-爬虫\数据解析\FZU.csv',encoding='utf-8')
df
```

```python
# 统计每个通知人的数量
notifiers = df['通知人'].value_counts()
# 计算比例
notifier_percents = notifiers / notifiers.sum() * 100
# 打印结果
print("通知人及其比例：")
print(notifier_percents)
```

```python
# 按通知人分组，计算各通知人每个通知平均附件下载次数
avg_downloads_by_notifier = df.groupby('通知人')['附件下载次数'].mean()
# 打印结果
print("各通知人每个通知的平均附件下载次数：")
print(avg_downloads_by_notifier)
# 可视化
plt.figure(figsize=(10, 6))
sns.barplot(x=avg_downloads_by_notifier.index, y=avg_downloads_by_notifier.values)
plt.title('各通知人每个通知的平均附件下载次数')
plt.xlabel('通知人')
plt.ylabel('附件下载次数')
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
plt.show()
#实践科与计划科与教研教改平均附件下载次数最多，
# 实践科：科研竞赛通知
# 计划科：选课开课通知
# 教研教：保研，数理金融实验班招生通知
# 这些部门都是发布都与学生学业息息相关
```

```python
import numpy as np

# 定义Logistic函数
def logistic(x, L, k, x0):
    return L / (1 + np.exp(-k * (x - x0)))

# 绘制Logistic函数
x = np.linspace(-10, 10, 100)
L = 1
k = 1
x0 = 0
y = logistic(x, L, k, x0)

plt.plot(x, y, label=f'L={L}, k={k}, x0={x0}')
plt.title('Logistic函数')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

# 尝试不同的L、k、x0
parameters = [
    (1, 1, 0),
    (2, 1, 0),
    (1, 2, 0),
    (1, 1, 2)
]

plt.figure(figsize=(10, 6))
for L, k, x0 in parameters:
    y = logistic(x, L, k, x0)
    plt.plot(x, y, label=f'L={L}, k={k}, x0={x0}')

plt.title('不同的L、k、x0的logistic函数')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
plt.show()


# k增加会使函数的斜率变缓
# x0增加会使函数斜率变抖
# l增加会使函数斜率和稳定值增加
```

```python
df['日期'] = pd.to_datetime(df['日期'])
# df['Month'] = df['Date'].dt.month
df['月'] = df['日期'].dt.to_period('M')
months = df.groupby('月').size()
print(months)
# 找出通知数最多的月份
most_month = months.idxmax()
print(f"通知最密集的月份是: {most_month}")
# 可视化
plt.figure(figsize=(20, 6))
sns.barplot(x=months.index.astype(str), y=months.values)
plt.title('每月通知数统计')
plt.xlabel('月份')
plt.ylabel('通知数')
plt.show()
```

```python
df['天'] = df['日期'].dt.to_period('D')
Days = df.groupby('天').size()
print(Days)
# 找出通知数最多的月份
most_days = Days.idxmax()
print(f"通知最密集的日期是: {most_days}")
# 可视化
plt.figure(figsize=(300, 10))
sns.barplot(x=Days.index.astype(str), y=Days.values)
plt.title('每天通知数统计')
plt.xlabel('日期')
plt.ylabel('通知数')
plt.show()
```
