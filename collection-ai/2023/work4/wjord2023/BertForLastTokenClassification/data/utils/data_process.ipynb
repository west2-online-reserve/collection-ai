### 数据分析

```python
import pandas as pd
from pandas import DataFrame

df: DataFrame = pd.read_json("../book_text.json", encoding="utf-8", lines=True)
```

```python
df.sort_values(by=["book_name", "chapter_index"], inplace=True)
df.reset_index(drop=True, inplace=True)
df
```

```python
def add_newline(text: str) -> str:
    if text[-1] != "\n":
        return text + "\n"
    else:
        return text


df["book_text"] = df["book_text"].str.strip()
df["book_text"] = df["book_text"].str.replace("　　", "")
df["book_text"] = df["book_text"].replace("", pd.NA)
df.dropna(subset=["book_text"], inplace=True)
df["book_text"] = df["book_text"].apply(add_newline)
df["book_text"].str.slice(-5)
```

```python
df.sample(10)
```

```python
def del_begin_punctuation(text: str) -> str:
    if text != "":
        if text[0] in "，。！？；:！：，。、":
            text = text[1:]
    return text


df["chapter_name"] = df["chapter_name"].apply(del_begin_punctuation)
df.sample(5)
```

```python
df["chapter_name"].sample(10)
```

```python
def del_useless_content(text: str) -> str:
    try:
        if text.startswith('【'):
            text = text.split('】', 1)[1]
        if text.startswith('（'):
            text = text.split('）', 1)[1]
        if text.endswith('】'):
            text = text.split('【', 1)[0]
        if text.endswith('）'):
            text = text.split('（', 1)[0]
        if '求订阅' in text:
            text = text.split('（', 1)[0]
            text = text.split('【', 1)[0]
            print(text)
        if text.startswith('('):
            text = text.split(')', 1)[1]
        if text.endswith(')'):
            text = text.split('(', 1)[0]
    except IndexError: 
        print(text)
    return text

df['chapter_name'] = df['chapter_name'].apply(del_useless_content)
df['chapter_name'].sample(5)
```

```python
df['chapter_name'].sample(20)
```

```python
def split_book_text(text: str, max_length=256):
    split_points = [0]  # 初始化拆分点列表，初始值为 0

    for i, char in enumerate(text):
        if i - split_points[-1] >= max_length and char == '\n':
            split_point = text.rfind('\n', split_points[-1], i)  # 找到最近的一个换行符
            if split_point == -1:
                split_point = i
            split_points.append(split_point)
                
    split_text = [text[split_points[i]:split_points[i+1]] for i in range(len(split_points)-1)]  # 使用拆分点将文本拆分为多个片段
    return split_text
    
def recursive_split(text, max_length=256):
    # 递归拆分每个片段，直到所有片段都不超过指定长度，并且至少拆分出一个非空列表
    final_text = []
    split_text = split_book_text(text, max_length)
    for part in split_text:
        if len(part) > max_length:
            sub_parts = split_book_text(part, max_length)
            if sub_parts:
                final_text.extend(sub_parts)
        elif part:
            final_text.append(part)

    return final_text

df['split_book_text'] = df['book_text'].apply(recursive_split)
```

```python
df['split_book_text'] = df['split_book_text'].str.strip()
df['split_book_text'] = df['split_book_text'].apply(add_newline)
```

```python
import numpy as np

def normal_split(text: str) -> str:
    text_len = len(text)
    mean = text_len // 3 # 设置正态分布的均值
    std = text_len // 6 # 设置正态分布的标准差
    split_point = int(np.random.normal(mean, std))
    split_point = max(1, min(split_point, text_len - 1))
    return text[:split_point], text[split_point:]

df['split_book_text'] = df['split_book_text'].astype(str)
df['split_book_text_1'], df['split_book_text_2'] = zip(*df['split_book_text'].apply(normal_split))
```
