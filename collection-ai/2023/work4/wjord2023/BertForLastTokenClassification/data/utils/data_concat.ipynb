```python
import pandas as pd

df = pd.read_json('../data_pre.json', lines=True)
df
```

```python
import numpy as np


# 50%的概率将书名设置为'无'
random_values = np.random.rand(df.shape[0])
df['book_name'] = np.where(random_values > 0.5, '无', df['book_name'])
# 50%的概率将类型设置为'无'
random_values = np.random.rand(df.shape[0])
df['book_type'] = np.where(random_values > 0.5, '无', df['book_type'])
# 50%的概率将章节名设置为'无'
random_values = np.random.rand(df.shape[0])
df['chapter_name'] = np.where(random_values > 0.7, '无', df['chapter_name'])

df
```

```python
data = pd.DataFrame(columns=['sentence_1', 'sentence_2'])
data
```

```python
data['sentence_1'] = df.apply(lambda row: '书名：' + row['book_name'] + '，类型：' + row['book_type'] + '，章节名：' + row['chapter_name'] + '，内容：' + row['split_book_text_1'], axis=1)
data['sentence_2'] = df['split_book_text_2']
data
```

```python
data.to_json('../data.json', orient='records', lines=True, force_ascii=False)
```
